# OpenML Arborration Regression Permutation Benchmark

Regression-only OpenML benchmark for the target-permutation-calibrated Arborration selector.

This notebook compares `all_features` against `selected_features` from `arborration.select_features_for_task` with `task="regression"`. The selector configuration exposes `n_target_permutations`, `n_permutation_refits`, and `permutation_scope` so the null-calibrated method can be tested directly.

```python
from __future__ import annotations

import sys
import time
import warnings
from pathlib import Path

import numpy as np
import pandas as pd

from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

REPO_ROOT = Path.cwd().resolve().parent if Path.cwd().name == "benchmarking" else Path.cwd().resolve()
SRC_DIR = REPO_ROOT / "src"
if str(SRC_DIR) not in sys.path:
    sys.path.insert(0, str(SRC_DIR))

from arborration import select_features_for_task  # noqa: E402

warnings.filterwarnings("ignore", category=UserWarning)
pd.set_option("display.max_columns", 200)
pd.set_option("display.max_rows", 200)

CACHE_DIR = REPO_ROOT / "benchmarking" / "openml_cache"
CACHE_DIR.mkdir(parents=True, exist_ok=True)
```

## Configuration

The defaults are intentionally smoke-sized. Increase task counts, repeats, selector trees, and model trees after confirming the workflow runs in your environment.


```python
CONFIG = {
    "random_state": 42,
    "n_repeats": 2,
    "test_size": 0.30,

    # Leave empty to auto-select regression tasks from OpenML metadata.
    "manual_regression_task_ids": [],
    "n_regression_tasks": 5,

    # Metadata filters before loading data.
    "min_features": 10,
    "max_features": 1000,
    "min_rows": 100,
    "max_rows": 10000,

    # Local encoded matrix filters after preprocessing.
    "min_encoded_features": 5,
    "max_encoded_features": 3000,
    "max_categorical_levels": 50,
    "drop_constant_features": True,

    "model_params": {
        "n_estimators": 200,
        "max_depth": None,
        "n_jobs": -1,
    },
    "selector_params": {
        "ntrees": 100,
        "sample_size": 256,
        "ndim": 2,
        "max_iter": 10,
        "n_refits": 1,
        "target_draw_probability": 0.25,
        "regression_target_bins": 10,
        "n_target_permutations": 5,
        "n_permutation_refits": 1,
        "permutation_scope": "borderline",
        "permutation_usage_margin": 1,
        "min_permutation_adjusted_usage": 0.0,
        "min_features_to_keep": 1,
        "max_refit_usage_fraction_for_removal": 0.0,
        "max_remove_fraction_per_iter": 0.10,
        "verbose": False,
    },
}
```

## OpenML Task Plan


```python
def _first_existing_column(df, candidates):
    for col in candidates:
        if col in df.columns:
            return col
    return None


def _numeric_column(df, candidates):
    col = _first_existing_column(df, candidates)
    if col is None:
        return pd.Series(np.nan, index=df.index)
    return pd.to_numeric(df[col], errors="coerce")


def load_openml_task_metadata(cache_path=CACHE_DIR / "openml_tasks_metadata.csv", *, refresh=False):
    if cache_path.exists() and not refresh:
        return pd.read_csv(cache_path)

    import openml

    tasks = openml.tasks.list_tasks(output_format="dataframe")
    tasks.columns = [str(c) for c in tasks.columns]
    tasks = tasks.reset_index()
    tasks.to_csv(cache_path, index=False)
    return tasks


def normalize_openml_task_metadata(tasks):
    tasks = tasks.copy()
    tasks.columns = [str(c) for c in tasks.columns]

    tid_col = _first_existing_column(tasks, ["tid", "task_id", "TaskID", "index"])
    did_col = _first_existing_column(tasks, ["did", "dataset_id", "DatasetID", "source_data"])
    name_col = _first_existing_column(tasks, ["name", "dataset", "dataset_name"])
    task_type_col = _first_existing_column(tasks, ["task_type", "TaskType"])
    target_col = _first_existing_column(tasks, ["target_feature", "target_name", "TargetFeature"])

    if tid_col is None or task_type_col is None:
        raise RuntimeError(f"OpenML task metadata is missing required columns: {tasks.columns.tolist()}")

    out = pd.DataFrame({
        "tid": pd.to_numeric(tasks[tid_col], errors="coerce"),
        "did": pd.to_numeric(tasks[did_col], errors="coerce") if did_col else np.nan,
        "dataset": tasks[name_col].astype(str) if name_col else "",
        "openml_task_type": tasks[task_type_col].astype(str),
        "target_feature": tasks[target_col].astype(str) if target_col else "",
        "n_features": _numeric_column(tasks, ["NumberOfFeatures", "number_features", "n_features"]),
        "n_rows": _numeric_column(tasks, ["NumberOfInstances", "number_instances", "n_instances"]),
    })
    task_type = out["openml_task_type"].str.lower()
    out["task"] = np.select([task_type.str.contains("regression")], ["regression"], default="other")
    return out.dropna(subset=["tid"]).drop_duplicates("tid")


def build_task_plan(refresh_metadata=False):
    rows = []
    for tid in CONFIG["manual_regression_task_ids"]:
        rows.append({"tid": int(tid), "task": "regression", "dataset": f"manual_task_{tid}"})
    if rows:
        return pd.DataFrame(rows)

    tasks = normalize_openml_task_metadata(load_openml_task_metadata(refresh=refresh_metadata))
    tasks = tasks[
        (tasks["task"] == "regression")
        & tasks["n_features"].between(CONFIG["min_features"], CONFIG["max_features"])
        & tasks["n_rows"].between(CONFIG["min_rows"], CONFIG["max_rows"])
    ].copy()
    return tasks.sort_values(["n_features", "n_rows"], ascending=False).head(CONFIG["n_regression_tasks"])


task_plan = build_task_plan(refresh_metadata=False)
task_plan.reindex(columns=["task", "tid", "dataset", "n_rows", "n_features"])
```

## Loading And Preprocessing


```python
def load_openml_task_xy(tid):
    import openml

    task = openml.tasks.get_task(int(tid))
    dataset = task.get_dataset()
    target_name = getattr(task, "target_name", None) or getattr(dataset, "default_target_attribute", None)
    if target_name is None:
        raise ValueError(f"Could not determine target for OpenML task {tid}.")

    X, y, _, _ = dataset.get_data(target=target_name, dataset_format="dataframe")
    X = X.copy()
    X.columns = [str(c) for c in X.columns]
    y = pd.Series(y, name="target")
    meta = {
        "tid": int(tid),
        "dataset_id": getattr(dataset, "dataset_id", np.nan),
        "dataset": getattr(dataset, "name", f"openml_task_{tid}"),
        "target_name": str(target_name),
    }
    return X, y, meta


def preprocess_xy(X_raw, y_raw):
    X = X_raw.copy().replace([np.inf, -np.inf], np.nan)
    X.columns = [str(c) for c in X.columns]
    y = pd.to_numeric(pd.Series(y_raw).copy(), errors="coerce")
    ok = y.notna()
    X = X.loc[ok].reset_index(drop=True)
    y = y.loc[ok].reset_index(drop=True).astype(float)

    numeric_parts = []
    categorical_parts = []
    dropped_high_cardinality = []

    for col in X.columns:
        series = X[col]
        if pd.api.types.is_bool_dtype(series) or pd.api.types.is_numeric_dtype(series):
            numeric_parts.append(pd.to_numeric(series, errors="coerce").astype(float).to_frame(col))
            continue

        if series.nunique(dropna=True) <= CONFIG["max_categorical_levels"]:
            categorical_parts.append(series.astype("category").to_frame(col))
        else:
            dropped_high_cardinality.append(col)

    parts = []
    if numeric_parts:
        X_num = pd.concat(numeric_parts, axis=1)
        X_num = X_num.fillna(X_num.median(axis=0, numeric_only=True)).fillna(0.0)
        parts.append(X_num.astype(float))
    if categorical_parts:
        X_cat = pd.concat(categorical_parts, axis=1)
        parts.append(pd.get_dummies(X_cat, dummy_na=True, dtype=float))

    if not parts:
        raise ValueError("No usable features after preprocessing.")

    X_out = pd.concat(parts, axis=1)
    X_out.columns = [str(c) for c in X_out.columns]
    if CONFIG["drop_constant_features"]:
        keep = X_out.nunique(dropna=False) > 1
        X_out = X_out.loc[:, keep]

    if not CONFIG["min_encoded_features"] <= X_out.shape[1] <= CONFIG["max_encoded_features"]:
        raise ValueError(f"Encoded feature count {X_out.shape[1]} outside configured bounds.")

    return X_out.reset_index(drop=True), y.reset_index(drop=True), {
        "n_dropped_high_cardinality": len(dropped_high_cardinality),
        "dropped_high_cardinality": dropped_high_cardinality,
    }
```

## Benchmark Helpers


```python
def build_model(seed):
    params = dict(CONFIG["model_params"])
    params["random_state"] = seed
    return RandomForestRegressor(**params)


def score_predictions(y_true, pred):
    rmse = float(np.sqrt(mean_squared_error(y_true, pred)))
    r2 = r2_score(y_true, pred)
    return {"rmse": rmse, "r2": r2, "primary_metric": r2}


def make_feature_sets(X_train, X_test, y_train, *, seed):
    feature_sets = {
        "all_features": {
            "X_train": X_train,
            "X_test": X_test,
            "selected_features": list(X_train.columns),
            "selector_method": "all_features",
            "selector_elapsed_sec": 0.0,
            "selector_error": None,
        }
    }

    try:
        t0 = time.perf_counter()
        X_train_selected, selector_result = select_features_for_task(
            X_train,
            y_train,
            task="regression",
            random_state=seed,
            **CONFIG["selector_params"],
        )
        elapsed = time.perf_counter() - t0
        selected = selector_result["selected_features"]
        feature_sets["selected_features"] = {
            "X_train": X_train_selected,
            "X_test": X_test[selected],
            "selected_features": selected,
            "selector_method": selector_result.get("method", "select_features_for_task"),
            "selector_elapsed_sec": elapsed,
            "selector_error": None,
        }
    except Exception as exc:
        feature_sets["selected_features"] = {
            "X_train": X_train,
            "X_test": X_test,
            "selected_features": list(X_train.columns),
            "selector_method": "select_features_for_task",
            "selector_elapsed_sec": np.nan,
            "selector_error": repr(exc),
        }

    return feature_sets
```

## Run Benchmark


```python
def run_benchmark(task_plan):
    rows = []
    for _, task_row in task_plan.iterrows():
        tid = int(task_row["tid"])
        try:
            X_raw, y_raw, openml_meta = load_openml_task_xy(tid)
            X, y, preprocess_meta = preprocess_xy(X_raw, y_raw)
        except Exception as exc:
            print(f"[WARN] Skipping task {tid}: {exc}")
            continue

        print(f"\n=== regression task={tid} dataset={openml_meta['dataset']} rows={len(X)} features={X.shape[1]} ===")

        for repeat in range(CONFIG["n_repeats"]):
            seed = CONFIG["random_state"] + repeat
            try:
                X_train, X_test, y_train, y_test = train_test_split(
                    X,
                    y,
                    test_size=CONFIG["test_size"],
                    random_state=seed,
                )
            except Exception as exc:
                print(f"[WARN] Split failed for task {tid} repeat={repeat}: {exc}")
                continue

            X_train = X_train.reset_index(drop=True)
            X_test = X_test.reset_index(drop=True)
            y_train = y_train.reset_index(drop=True)
            y_test = y_test.reset_index(drop=True)
            feature_sets = make_feature_sets(X_train, X_test, y_train, seed=seed)

            selected_count = len(feature_sets["selected_features"]["selected_features"])
            print(f"repeat={repeat}: selected {selected_count}/{X.shape[1]} features")

            for feature_set, payload in feature_sets.items():
                train_time = np.nan
                infer_time = np.nan
                metrics = {}
                model_error = None
                try:
                    model = build_model(seed)
                    t0 = time.perf_counter()
                    model.fit(payload["X_train"], y_train)
                    train_time = time.perf_counter() - t0
                    t0 = time.perf_counter()
                    pred = model.predict(payload["X_test"])
                    infer_time = time.perf_counter() - t0
                    metrics = score_predictions(y_test, pred)
                except Exception as exc:
                    model_error = repr(exc)

                rows.append({
                    "task": "regression",
                    "tid": tid,
                    "dataset": openml_meta["dataset"],
                    "repeat": repeat,
                    "feature_set": feature_set,
                    "selector_method": payload["selector_method"],
                    "n_original_features": X.shape[1],
                    "n_selected_features": len(payload["selected_features"]),
                    "selector_elapsed_sec": payload["selector_elapsed_sec"],
                    "selector_error": payload["selector_error"],
                    "train_time_sec": train_time,
                    "infer_time_sec": infer_time,
                    "model_error": model_error,
                    **metrics,
                    **preprocess_meta,
                })
    return pd.DataFrame(rows)


results = run_benchmark(task_plan)
results.head()
```

## Selected Features Versus All Features


```python
keys = ["task", "tid", "dataset", "repeat"]
baseline = (
    results.loc[results["feature_set"] == "all_features", keys + ["primary_metric"]]
    .rename(columns={"primary_metric": "primary_metric_all_features"})
)
comparison = results.merge(baseline, on=keys, how="left")
comparison["delta_vs_all_features"] = comparison["primary_metric"] - comparison["primary_metric_all_features"]
comparison = comparison[comparison["feature_set"] == "selected_features"].copy()

summary = (
    comparison.groupby(["task", "tid", "dataset"], dropna=False)
    .agg(
        mean_primary_metric=("primary_metric", "mean"),
        mean_delta_vs_all_features=("delta_vs_all_features", "mean"),
        mean_selected_features=("n_selected_features", "mean"),
        original_features=("n_original_features", "first"),
        mean_selector_time_sec=("selector_elapsed_sec", "mean"),
        selector_errors=("selector_error", lambda s: sorted({x for x in s.dropna()})),
        n_runs=("primary_metric", "count"),
    )
    .reset_index()
    .sort_values(["task", "mean_delta_vs_all_features"], ascending=[True, False])
)
summary
```

```python
output_dir = REPO_ROOT / "benchmarking" / "results"
output_dir.mkdir(parents=True, exist_ok=True)
results_path = output_dir / "openml_arborration_regression_permutation_results.csv"
comparison_path = output_dir / "openml_arborration_regression_permutation_comparison.csv"
summary_path = output_dir / "openml_arborration_regression_permutation_summary.csv"

results.to_csv(results_path, index=False)
comparison.to_csv(comparison_path, index=False)
summary.to_csv(summary_path, index=False)

print(f"Saved results to {results_path}")
print(f"Saved comparison to {comparison_path}")
print(f"Saved summary to {summary_path}")
```
